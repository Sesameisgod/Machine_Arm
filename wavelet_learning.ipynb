```python
# 引入 library
import matplotlib.pyplot as plt
import numpy as np
import pywt
```

```python
# 產生一正弦波
x = np.linspace(-np.pi, 5*np.pi, 600, endpoint=True)
y = np.sin(x)
plt.plot(y)
```

```python
# 使用多階小波包轉換
# 小波包基底: db5
# 層數: 4
# y = np.sin(x)
y = np.linspace(0,10,260)
wp = pywt.WaveletPacket(data=y, wavelet='db5', mode='symmetric', maxlevel=4)
```

```python
# 印出所有節點的名稱
for i in [1,2,3,4]:
    print([node.path for node in wp.get_level(i, 'natural')])
```
